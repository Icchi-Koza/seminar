```python
import pandas as pd
```

```python
dat1=pd.read_excel('ma-japan2.xlsx','Sheet1',index_col=None)
dat2=pd.read_excel('ma-japan3.xlsx','Sheet1',index_col=None)
dat3=pd.read_excel('ma-japan4.xlsx','Sheet1',index_col=None)
dat4=pd.read_excel('ma-japan5.xlsx','Sheet1',index_col=None)
```

```python
dat_ma_jap=pd.concat([dat1,dat2,dat3,dat4])
```

```python
dat_ma_jap['発表日']=pd.to_datetime(dat_ma_jap['発表日'],format='%m/%d/%Y')
dat_ma_jap=dat_ma_jap.sort_values('発表日').reset_index(drop=True)
```

```python
dat_ma_jap.to_csv('MA-JAPAN.csv',index=None,encoding='UTF-8')
```
