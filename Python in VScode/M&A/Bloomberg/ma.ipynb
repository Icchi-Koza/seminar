```python
import pandas as pd
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
dat1=pd.read_excel('ma-us1.xlsx','Sheet1',index_col=None)
dat2=pd.read_excel('ma-us2.xlsx','Sheet1',index_col=None)
dat3=pd.read_excel('ma-us3.xlsx','Sheet1',index_col=None)
dat4=pd.read_excel('ma-us4.xlsx','Sheet1',index_col=None)
dat5=pd.read_excel('ma-us5.xlsx','Sheet1',index_col=None)
dat6=pd.read_excel('ma-us6.xlsx','Sheet1',index_col=None)
dat7=pd.read_excel('ma-us7.xlsx','Sheet1',index_col=None)
dat8=pd.read_excel('ma-us8.xlsx','Sheet1',index_col=None)
dat9=pd.read_excel('ma-us9.xlsx','Sheet1',index_col=None)
dat10=pd.read_excel('ma-us10.xlsx','Sheet1',index_col=None)
```

```python
dat_ma_us=pd.concat([dat1,dat2,dat3,dat4,dat5,dat6,dat7,dat8,dat9])
dat_ma_us['発表日']=pd.to_datetime(dat_ma_us['発表日'],format='%m/%d/%Y')
dat_ma_us=dat_ma_us.sort_values('発表日').reset_index(drop=True)
dat10['発表日']=pd.to_datetime(dat10['発表日'],format='%m/%d/%Y')
dat_ma_us=pd.merge(dat_ma_us,dat10,how='left',on=['発表日','ﾀｰｹﾞｯﾄ会社名','買収側の社名','発表時ﾃﾞｨｰﾙ総額(百万)'])
```

```python
dat_ma_us.to_csv('MA_US.csv',index=None,encoding='UTF-8')
```
